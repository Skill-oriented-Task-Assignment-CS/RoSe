```python
%%time
source_file_location =  "dataset/final_job.csv"
mapper_file_location =  "dataset/upwork_jobs_mapper.json"
df_intermediate,old_column_names,new_column_names = create_intermediate_representation(source_file_location, mapper_file_location )
```

```python
save_list( old_column_names, "lib/old_column_names_upwork_jobs.dat")
save_list( new_column_names, "lib/new_column_names_upwork_jobs.dat")
```

```python
df_source=df_intermediate
df_intermediate.head()
```

```python
#creating the metadata

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=df_source)
metadata.validate()
print(metadata)
# traning the model
epoch=5000
synthesizer = CTGANSynthesizer(metadata,epochs=epoch,verbose=True)
synthesizer.fit(df_source)
```
